From the below function, you can analyse different countries based on yearly increase in suicide rate and match them with the changes in the GDP.

```python
import matplotlib.pyplot as plt
def year_wise_suicide_rate_countryWise(country):
    year = set()
    age_grup = set()
    a=0
    for j,i in df.iterrows():
        if i["country"]==country:
            a=1
            year.add(i["year"])
            age_grup.add(i["age"])
        elif a==1:
            a=0
            break
    temp=0
    country_year_df = pd.DataFrame(columns=["year","suicide_rate","GDP_per_capita"])
    index = 0
    for cnt_year in year:
        temp = 0
        country_year_df.loc[index] = [cnt_year,0,0]
        b = 0
        for j,i in df.iterrows():
            if i["year"] == cnt_year:
                if a==0:
                    a=1
                    b+=1
                    country_year_df = country_year_df.set_value(index,"GDP_per_capita",i["gdp_per_capita ($)"])
                    temp+=i["suicides/100k pop"]
                else:
                    temp+=i["suicides/100k pop"]
                    b+=1
            elif a==1:
                a=0
                country_year_df = country_year_df.set_value(index,"suicide_rate",temp/b)
                print(b)
                index+=1
                break
    return country_year_df
df_2 = year_wise_suicide_rate_countryWise("Albania")
print(df_2)
fig, ax = plt.subplots()
ax.plot(df_2["year"], df_2["suicide_rate"])
ax.set(xlabel="year", ylabel="suicide_rate",title="Albania")
ax.grid()
plt.show()
# sns.set(rc={'figure.figsize':(11.7,8.27)})
# ax = sns.barplot(x="year", y="suicide_rate", data=df_2)
# ax
```

Since there are too many countries to analyse, so I tried a simple way to distribute these countries into continents which are only 6 in number. This will help us to get a good understanding on a large basis.

```python
cont_df = pd.read_csv(r"continent.csv")
up_df = pd.concat([df, cont_df], axis=1)
```

```python
up_df.head(10)
```

Now we try to get the continents to work. So, instead of countries, we are now looking for continents.

```python
def continent_analyse(cnt):
    age_grup = set()
    a=0
    for j,i in up_df.iterrows():
        if i["continent"]==cnt:
            age_grup.add(i["age"])
    continent_age_df = pd.DataFrame(columns=["age_grup","suicide_rate","GDP_per_capita"])
    index = 0
    for cnt_age in age_grup:
        sui_temp=0
        gdp_temp=0
        continent_age_df.loc[index] = [cnt_age,0,0]
        b = 0
        for j,i in up_df.iterrows():
            if i["continent"]==cnt:
                if i["age"] == cnt_age:
                    sui_temp+=i["suicides/100k pop"]
                    gdp_temp+=i["gdp_per_capita ($)"]
                #print(gdp_temp)
                    b+=1
        continent_age_df = continent_age_df.set_value(index,"suicide_rate",sui_temp/b)
        continent_age_df = continent_age_df.set_value(index,"GDP_per_capita",gdp_temp/b)
        index+=1
    return continent_age_df

def bar_continent(temp_df):
    sns.set(rc={'figure.figsize':(8,6)})
    ax = sns.barplot(x="age_grup", y="suicide_rate", data=temp_df)
df3 = continent_analyse("Europe")
bar_continent(df3)
df3
    
    
```

From the above chart, we can conclude that in EUROPE, their are more chances of a 75+ year person committing suicide. But we should take this graph as number of suicides per 100k population, not the actual number of suicides because the population of 75+ years people is less, thats why the rate is high. While there are approximately equal chances of people of the age group "35-54" and "55-74" years committing suicides.

```python
df3['suicide_rate'] = df3['suicide_rate'].astype(float)
ax = sns.boxplot(x=df3["suicide_rate"])
```
